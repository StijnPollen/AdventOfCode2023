```python
file = open(r"C:\Users\Pollen\Documents\Afstudeerstage\AdventOfCode\day3input.txt")

grid = []
#Read each line from the file
for line in file.readlines():
    row = []
    #Iterate through each character in the line
    for column in line:
        row.append(column)
    #Add the row to the grid
    grid.append(row)

#Initialize the list to store found coordinates
found_coords = []

#Check if character has non-digit, non-dot symbols horizontally.
for i in range(len(grid)):
    for j in range(len(grid[0])-1):
        found_symbol = False
        if grid[i][j].isdigit():
            #Check to the right
            if j + 2 < len(grid[0]) and not (grid[i][j + 1].isdigit() or grid[i][j + 1] == "."):
                found_symbol = True
            #Check to the left
            if j - 1 >= 0 and not (grid[i][j - 1].isdigit() or grid[i][j - 1] == "."):
                found_symbol = True

            #Check below
            if i + 1 < len(grid) and not (grid[i + 1][j].isdigit() or grid[i + 1][j] == "."):
                found_symbol = True
            #Check above
            if i - 1 >= 0 and not (grid[i - 1][j].isdigit() or grid[i - 1][j] == "."):
                found_symbol = True

            #Check left below
            if i + 1 < len(grid) and j - 1 >= 0 and not (grid[i + 1][j - 1].isdigit() or grid[i + 1][j - 1] == "."):
                found_symbol = True
            #Check left above
            if i - 1 >= 0 and j - 1 >= 0 and not (grid[i - 1][j - 1].isdigit() or grid[i - 1][j - 1] == "."):
                found_symbol = True
            #Check right below
            if i + 1 < len(grid) and j + 2 < len(grid[0]) and not (grid[i + 1][j + 1].isdigit() or grid[i + 1][j + 1] == "."):
                found_symbol = True
            #Check right above
            if i - 1 >= 0 and j + 2 < len(grid[0]) and not (grid[i - 1][j + 1].isdigit() or grid[i - 1][j + 1] == "."):
                found_symbol = True

        #Find first digit of number
        if found_symbol:
            if j - 2 >= 0 and grid[i][j - 2].isdigit() and grid[i][j - 1].isdigit():
                found_coords.append((i, j - 2))
            elif j - 1 >= 0 and grid[i][j - 1].isdigit():
                found_coords.append((i, j - 1))
            else:
                found_coords.append((i, j))

#Remove dupes
found_coords_nodupes = list(dict.fromkeys(found_coords))
#Turn coords into numbers and add them to total_value
total_value = 0
for coords in found_coords_nodupes:
    i, j = coords
    #Check if the number after the coords in grid[i][j] is also a digit
    if j + 2 < len(grid[0]) and grid[i][j + 1].isdigit() and grid[i][j + 2].isdigit():
        #If so, turn it into an integer and add it to total_value
        num_str = int(grid[i][j] + grid[i][j + 1] + grid[i][j + 2])
    elif j + 1 < len(grid[0]) and grid[i][j + 1].isdigit():
        num_str = int(grid[i][j] + grid[i][j + 1])
    else:
        num_str = int(grid[i][j])
    total_value += num_str

print(total_value)
        
file.close()
```

```python
#Check for gears
gear_coords = []
for i in range(len(grid)):
    for j in range(len(grid[0])-1):
        if grid[i][j] == "*":
            gear_coords.append((i, j))

#Check for numbers
number_coords = []
for i in range(len(grid)):
    for j in range(len(grid[0])-1):
        if grid[i][j].isdigit():
            number_coords.append((i, j))

#Store the total multiplied value
total_value = 0

#Check for all gear coords if 2 numbers are close
for gear_coord in gear_coords:
    gear_count = 0
    i, j = gear_coord
    adjacent_numbers = []

    #Check horizontally to the right
    if (i, j + 1) in number_coords:
        gear_count += 1
        #Find first digit of number
        adjacent_numbers.append((i, j + 1))

    #Check horizontally to the left
    if (i, j - 1) in number_coords:
        gear_count += 1
        #Find first digit of number
        if j - 3 >= 0 and grid[i][j - 3].isdigit() and grid[i][j - 2].isdigit():
            adjacent_numbers.append((i, j - 3))
        elif j - 2 >= 0 and grid[i][j - 2].isdigit():
            adjacent_numbers.append((i, j - 2))
        else:
            adjacent_numbers.append((i, j - 1))

    #Check vertically below
    if (i + 1, j) in number_coords:
        gear_count += 1
        #Find first digit of number
        if j - 2 >= 0 and grid[i + 1][j - 2].isdigit() and grid[i + 1][j - 1].isdigit():
            adjacent_numbers.append((i + 1, j - 2))
        elif j - 1 >= 0 and grid[i + 1][j - 1].isdigit():
            adjacent_numbers.append((i + 1, j - 1))
        else:
            adjacent_numbers.append((i + 1, j))

    #Check vertically above
    if (i - 1, j) in number_coords:
        gear_count += 1
        #Find first digit of number
        if j - 2 >= 0 and grid[i - 1][j - 2].isdigit() and grid[i - 1][j - 1].isdigit():
            adjacent_numbers.append((i - 1, j - 2))
        elif j - 1 >= 0 and grid[i - 1][j - 1].isdigit():
            adjacent_numbers.append((i - 1, j - 1))
        else:
            adjacent_numbers.append((i - 1, j))

    #Check diagonally right below
    if (i + 1, j + 1) in number_coords and (i + 1, j) not in number_coords:
        gear_count += 1
        #Find first digit of number
        adjacent_numbers.append((i + 1, j + 1))

    #Check diagonally left below
    if (i + 1, j - 1) in number_coords  and (i + 1, j) not in number_coords:
        gear_count += 1
        #Find first digit of number
        if j - 3 >= 0 and grid[i + 1][j - 3].isdigit() and grid[i + 1][j - 2].isdigit():
            adjacent_numbers.append((i + 1, j - 3))
        elif j - 2 >= 0 and grid[i + 1][j - 2].isdigit():
            adjacent_numbers.append((i + 1, j - 2))
        else:
            adjacent_numbers.append((i + 1, j - 1))

    #Check diagonally right above
    if (i - 1, j + 1) in number_coords  and (i - 1, j) not in number_coords:
        gear_count += 1
        #Find first digit of number
        adjacent_numbers.append((i - 1, j + 1))

    #Check diagonally left above
    if (i - 1, j - 1) in number_coords  and (i - 1, j) not in number_coords:
        gear_count += 1
        #Find first digit of number
        if j - 3 >= 0 and grid[i - 1][j - 3].isdigit() and grid[i - 1][j - 2].isdigit():
            adjacent_numbers.append((i - 1, j - 3))
        elif j - 2 >= 0 and grid[i - 1][j - 2].isdigit():
            adjacent_numbers.append((i - 1, j - 2))
        else:
            adjacent_numbers.append((i - 1, j - 1))
            
    #If adjacent_numbers has 2 values, get the numbers at the coords and multiply them.
    if len(adjacent_numbers) == 2:
        num_str = []
        for coords in adjacent_numbers:
            i, j = coords
            #Check if the number after the coords in grid[i][j] is also a digit
            if j + 2 < len(grid[0]) and grid[i][j + 1].isdigit() and grid[i][j + 2].isdigit():
                #If so, turn it into an integer and add it to num_str
                num_str.append(int(grid[i][j] + grid[i][j + 1] + grid[i][j + 2]))
            elif j + 1 < len(grid[0]) and grid[i][j + 1].isdigit():
                num_str.append(int(grid[i][j] + grid[i][j + 1]))
            else:
                num_str.append(int(grid[i][j]))
        total_value += num_str[0] * num_str[1]
        
print(total_value)

```
